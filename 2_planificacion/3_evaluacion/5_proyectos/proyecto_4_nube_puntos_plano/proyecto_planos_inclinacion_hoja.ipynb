# **Anlisis de imgenes, anlisis geomtrico y aplicaciones agropecuarias**.  


Te mostrar paso a paso cmo:

1. **Graficar el plano** $2x + 3y + 4z = 0$ en 3D con **matplotlib**.
2. Dibujar su **vector normal**.
3. Calcular el **ngulo de inclinacin** entre ese plano y el plano $xy$.

---



# Fundamento matemtico

El plano est dado por:
$$
2x + 3y + 4z = 0
$$

* Su **vector normal** es $\vec{n} = (2, 3, 4)$.
* El plano $xy$ tiene normal $\vec{n}_{xy} = (0, 0, 1)$.



El **ngulo de inclinacin** $\theta$ entre ambos planos es el ngulo entre sus normales:

$$
\cos(\theta) = \frac{\vec{n} \cdot \vec{n}*{xy}}{||\vec{n}|| , ||\vec{n}*{xy}||}
$$



$$
\Rightarrow \theta = \arccos\left( \frac{4}{\sqrt{2^2 + 3^2 + 4^2}} \right)
$$




# Cdigo completo en Python


```python
import numpy as np
import matplotlib.pyplot as plt

# Coeficientes del plano
a, b, c, d = 0, 0, 1, 0

# Crear una cuadrcula para x e y
x = np.linspace(-5, 5, 20)
y = np.linspace(-5, 5, 20)
X, Y = np.meshgrid(x, y)

# Resolver z del plano: 2x + 3y + 4z = 0  z = -(2x + 3y)/4
Z = -(a*X + b*Y)/c

# Grfico 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Graficar el plano
ax.plot_surface(X, Y, Z, alpha=0.5, color='skyblue')

# Vector normal (2,3,4) en el origen
ax.quiver(0, 0, 0, a, b, c, color='red', length=4, normalize=True)

# Etiquetas
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Plano 2x + 3y + 4z = 0 y su vector normal')

# Ajustar lmites visuales
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_zlim(-5, 5)

plt.savefig(r"C:\Users\marco\Downloads\angulo_plano.png")
plt.show()
```

```python

# Clculo del ngulo de inclinacin respecto al plano xy
n = np.array([a, b, c])
n_xy = np.array([0, 0, 1])

cos_theta = np.dot(n, n_xy) / (np.linalg.norm(n) * np.linalg.norm(n_xy))
theta_rad = np.arccos(cos_theta)
theta_deg = np.degrees(theta_rad)

print(f"ngulo de inclinacin del plano respecto al plano XY: {theta_deg:.2f}")

```

# **Uso de imgenes como nubes de puntos para determinar el ngulo de inclinacin de una hoja en ingeniera agropecuaria**



### **1. Contexto agropecuario**

En ingeniera agropecuaria, el **ngulo de inclinacin de una hoja** (o *leaf inclination angle*, LIA) es una medida clave en:

* **Fisiologa vegetal:** Permite estudiar la orientacin de las hojas respecto a la luz solar.
* **Modelado de radiacin:** Se usa en modelos de intercepcin de luz (como en ndices LAI o PAR).
* **Monitoreo de estrs hdrico o nutricional:** Las hojas tienden a inclinarse ms o menos segn el estado hdrico o la turgencia.
* **Ingeniera de precisin:** Ayuda a estimar biomasa o productividad con sensores pticos o cmaras RGB / multiespectrales.

---



# **2. Representacin de imgenes como nubes de puntos**

Una **nube de puntos (point cloud)** es un conjunto de puntos en un espacio 3D que representan la superficie de un objeto.


En este caso, una hoja puede capturarse como una superficie tridimensional donde cada punto tiene coordenadas $(x, y, z)$.



Hay dos formas de obtener estas nubes:

#### a) **Desde imgenes 2D**

1. Capturas una imagen RGB de la hoja.
2. Procesas la imagen con tcnicas de segmentacin (por ejemplo, con OpenCV) para aislarla del fondo.
3. Asignas coordenadas a cada pxel en funcin de su posicin $(x, y)$ y de una variable asociada (por ejemplo, intensidad o elevacin).
4. Generas una nube de puntos simulada, til para hojas planas.



#### b) **Desde sensores 3D o fotogrametra**

1. Usas cmaras estreo, sensores LiDAR o fotogrametra (con drones o cmaras multivista).
2. El software (como **Agisoft Metashape**, **Pix4D**, o libreras como **Open3D** o **PCL**) reconstruye la hoja en 3D.
3. Obtienes una nube de puntos real con coordenadas espaciales.

---



### **3. Clculo del ngulo de inclinacin de la hoja**

Una vez tienes la nube de puntos $(x_i, y_i, z_i)$, puedes estimar el **plano medio de la hoja**.
Matemticamente, el plano puede representarse como:

$$
ax + by + cz + d = 0
$$

Los coeficientes $(a, b, c)$ definen el **vector normal** $\vec{n} = (a, b, c)$ al plano de la hoja.

El **ngulo de inclinacin** $\theta$ respecto al plano horizontal se calcula como:

$$
\theta = \arccos\left(\frac{c}{\sqrt{a^2 + b^2 + c^2}}\right)
$$

* Si $\theta = 0$: la hoja est completamente horizontal.
* Si $\theta = 90$: la hoja est vertical.

```python
# ======================================================
# Visualizacin interactiva 3D con Plotly de una hoja
# e inclusin del plano ajustado con Open3D (RANSAC)
# ======================================================

import numpy as np
import open3d as o3d
import plotly.graph_objects as go

# 1. Cargar la nube de puntos
pcd = o3d.io.read_point_cloud("hoja_inclinada_realista.ply")

# Verificar carga
if len(pcd.points) == 0:
    raise ValueError(" La nube de puntos est vaca o la ruta del archivo es incorrecta.")

points = np.asarray(pcd.points)
print(f" Nube cargada con {points.shape[0]} puntos.")

# 2. Ajustar un plano a la nube (RANSAC)
plane_model, inliers = pcd.segment_plane(distance_threshold=0.001,
                                         ransac_n=3,
                                         num_iterations=1000)
[a, b, c, d] = plane_model

# Calcular ngulo de inclinacin respecto al eje Z
theta = np.degrees(np.arccos(c / np.sqrt(a**2 + b**2 + c**2)))
print(f"ngulo de inclinacin estimado: {theta:.2f}")

# 3. Preparar datos para el plano ajustado
x_vals = np.linspace(points[:, 0].min(), points[:, 0].max(), 20)
y_vals = np.linspace(points[:, 1].min(), points[:, 1].max(), 20)
xx, yy = np.meshgrid(x_vals, y_vals)
zz = (-a * xx - b * yy - d) / c  # Ecuacin del plano

# 4. Crear figura 3D con Plotly
fig = go.Figure()

# Nube de puntos (hoja)
fig.add_trace(go.Scatter3d(
    x=points[:, 0],
    y=points[:, 1],
    z=points[:, 2],
    mode='markers',
    marker=dict(size=2, color=points[:, 2], colorscale='Viridis', opacity=0.8),
    name='Nube de puntos (hoja)'
))

# Plano ajustado
fig.add_trace(go.Surface(
    x=xx, y=yy, z=zz,
    colorscale='Reds',
    opacity=0.4,
    name=f'Plano ajustado ( = {theta:.2f})'
))

# 5. Configuracin esttica
fig.update_layout(
    title=f"Visualizacin 3D - Hoja y Plano Ajustado ( = {theta:.2f})",
    scene=dict(
        xaxis_title='Eje X',
        yaxis_title='Eje Y',
        zaxis_title='Eje Z (altura)',
        aspectmode='data',
    ),
    width=900,
    height=700,
    template='plotly_white'
)

# 6. Mostrar figura interactiva
fig.show()
```
