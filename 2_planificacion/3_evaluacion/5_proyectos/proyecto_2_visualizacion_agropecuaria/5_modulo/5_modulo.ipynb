# Ejemplo de anlisis de imagenes para el congreso  

Este es un **script de Python** para segmentar y mostrar las **tres porciones del cultivo** (no maduro, maduro y enfermo) en forma de matriz de imgenes utilizando **OpenCV y Matplotlib**.



Este enfoque usa **segmentacin por color (espacio HSV)** y visualiza los resultados con `matplotlib.pyplot.subplot`.


```python
# ===============================================
# Segmentacin de imagen de cultivo de arroz
# Autor: Marco Julio Caas
# Contexto: Agro Caucasiano - OpenCV + Matplotlib
# ===============================================

import cv2
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Cargar la imagen ---
image_path = r"C:\Users\marco\Documentos\docencia\vector_geometry_and_linear_algebra\2_planificacion\0_programa_curso\cultivo_arroz.png"
img = cv2.imread(image_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# --- 2. Convertir a espacio de color HSV ---
hsv = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HSV)

# --- 3. Definir rangos de color para cada tipo de cultivo ---
# (ajusta los valores segn el anlisis con cv2.inRange)

# Cultivo no maduro (verde intenso)
lower_green = np.array([35, 50, 50])
upper_green = np.array([85, 255, 255])
mask_green = cv2.inRange(hsv, lower_green, upper_green)

# Cultivo madurado o cosechado (amarillo/dorado)
lower_yellow = np.array([15, 50, 50])
upper_yellow = np.array([35, 255, 255])
mask_yellow = cv2.inRange(hsv, lower_yellow, upper_yellow)

# Cultivo enfermo (marrn oscuro o grisceo)
lower_brown = np.array([5, 50, 20])
upper_brown = np.array([25, 255, 180])
mask_brown = cv2.inRange(hsv, lower_brown, upper_brown)

# --- 4. Aplicar las mscaras a la imagen original ---
segment_green = cv2.bitwise_and(img_rgb, img_rgb, mask=mask_green)
segment_yellow = cv2.bitwise_and(img_rgb, img_rgb, mask=mask_yellow)
segment_brown = cv2.bitwise_and(img_rgb, img_rgb, mask=mask_brown)

# --- 5. Mostrar la matriz de imgenes con Matplotlib ---
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

axes[0, 0].imshow(img_rgb)
axes[0, 0].set_title("Imagen original")

axes[0, 1].imshow(segment_green)
axes[0, 1].set_title("Cultivo no maduro (verde)")

axes[1, 0].imshow(segment_yellow)
axes[1, 0].set_title("Cultivo madurado / cosechado (amarillo)")

axes[1, 1].imshow(segment_brown)
axes[1, 1].set_title("Cultivo enfermo (marrn)")

for ax in axes.flat:
    ax.axis("off")

plt.tight_layout()
plt.savefig(r"C:\Users\marco\Downloads\descomposicion_imagen_cultivo.png")
plt.show()
```

## **Explicacin conceptual**

* Se transforma la imagen de **RGB  HSV**, porque el espacio HSV permite segmentar mejor los colores por matices (Hue).
* Se crean **mscaras binarias** (`cv2.inRange`) para cada tipo de color asociado al estado del cultivo.
* Luego se combinan las mscaras con la imagen original usando `cv2.bitwise_and`.
* Finalmente, se usa una **matriz de subgrficos** para visualizar las tres etapas del cultivo junto a la imagen original.
