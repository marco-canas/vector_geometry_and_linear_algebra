
# **3. Lectura, redimensionamiento y almacenamiento**

Cada imagen se procesar as:

1. Se lee con `PIL.Image` o `cv2`
2. Se convierte a RGB
3. Se redimensiona
4. Se almacena:

   * Imagen original procesada
   * Imagen aplanada (flatten) para ML clsico
   * Etiqueta (0 o 1 segn carpeta)



# **CDIGO PYTHON: Lectura, redimensionamiento y almacenamiento**


````python
import os # crear archivos leer directorios   
import numpy as np
from PIL import Image
````

````python

# --------------------------------------------------------------------
# 1. Rutas del dataset
# --------------------------------------------------------------------
root_dir = r"C:\Users\marco\OneDrive - Universidad de Antioquia\Documentos\2_recursos_ensenanza\3_vec_geo_and_linear_algebra\dataset_para_clasificacin_binaria"

subfolders = {
    "Curl_Virus": 0,      # etiqueta 0
    "Healthy_Leaf": 1     # etiqueta 1
}
````

````python

# --------------------------------------------------------------------
# 2. Tamao deseado para todas las imgenes
# --------------------------------------------------------------------
img_size = (128, 128)
````

````python

# --------------------------------------------------------------------
# 3. Estructuras para almacenar los datos
# --------------------------------------------------------------------
X_images = []      # imgenes procesadas
X_flat = []        # imgenes aplanadas para ML clsico
y_labels = []      # etiquetas (0 o 1)
````

````python

# --------------------------------------------------------------------
# 4. Recorrer subcarpetas, leer, convertir, redimensionar y almacenar
# --------------------------------------------------------------------
for folder_name, label in subfolders.items():
    folder_path = os.path.join(root_dir, folder_name)

    for filename in os.listdir(folder_path):
        if filename.lower().endswith(".jpg"):

            filepath = os.path.join(folder_path, filename)

            try:
                # Leer la imagen
                img = Image.open(filepath)

                # Convertir a RGB
                img = img.convert("RGB")

                # Redimensionar a tamao estndar
                img = img.resize(img_size)

                # Convertir a array numpy
                img_array = np.asarray(img)

                # Guardar imagen procesada
                X_images.append(img_array)

                # Guardar imagen aplanada (flatten)
                X_flat.append(img_array.flatten())

                # Guardar etiqueta (0 o 1)
                y_labels.append(label)

            except Exception as e:
                print(f"Error procesando {filepath}: {e}")

# --------------------------------------------------------------------
# 5. Convertir todo a arreglos numpy
# --------------------------------------------------------------------
X_images = np.array(X_images)
X_flat = np.array(X_flat)
y_labels = np.array(y_labels)
````

````python

from sklearn.utils import Bunch
import numpy as np

# --------------------------------------------------------------------
# Se asume que ya existen:
# X_images, X_flat, y_labels
# --------------------------------------------------------------------

# Nombres de etiquetas (importante para interpretabilidad)
target_names = np.array(["Curl_Virus", "Healthy_Leaf"])

# --------------------------------------------------------------------
# Crear el objeto tipo Bunch
# --------------------------------------------------------------------
dataset = Bunch(
    data = X_flat,             # imgenes aplanadas
    images = X_images,         # imgenes en formato (128, 128, 3)
    target = y_labels,         # vector de 0 y 1
    target_names = target_names,
    DESCR = "Dataset de hojas sanas y hojas con Curl Virus para clasificacin binaria"
)

# --------------------------------------------------------------------
# Resumen del dataset estilo MNIST
# --------------------------------------------------------------------
print("=== BUNCH CREADO CON XITO ===")
print("dataset.data shape: ", dataset.data.shape)
print("dataset.images shape:", dataset.images.shape)
print("dataset.target shape:", dataset.target.shape)
print("Clases:", dataset.target_names)
print("Descripcin:", dataset.DESCR)
````

En consecuencia, fomentar su aprendizaje desde los niveles iniciales hasta la educacin superior resulta indispensable para formar profesionales capaces de enfrentar los retos de la sociedad contempornea.



En el campo de la didctica de las matemticas, **Papert (1980, 1996)** haba anticipado esta necesidad al concebir las computadoras no como simples herramientas de clculo, sino como **objetos para pensar**, capaces de favorecer la construccin activa de significados matemticos a travs de la programacin. 

Esta visin se alinea con el constructivismo de **Piaget**, que resalta el valor de la experimentacin en el aprendizaje, y con la perspectiva socio-constructivista de **Vygotsky**, que otorga un papel central a las herramientas culturales en la mediacin del conocimiento.



En este marco, el uso de entornos como **Jupyter Notebooks** y lenguajes como **Python, LaTeX, Markdown e ingls** no debe interpretarse como una rivalidad con la enseanza tradicional de la matemtica, sino como una **ampliacin de los registros de representacin** que describi **Duval (1993, 2006)**. 

Segn su teora semitica, el aprendizaje significativo se potencia cuando el estudiante logra traducir y articular distintos sistemas de representacin: grfico, algebraico, numrico, verbal y, en nuestro tiempo, digital.



De igual manera, el **Enfoque Ontosemitico de Godino (2002, 2017)** muestra que los significados matemticos emergen de prcticas diversas, lo que justifica la incorporacin de estos lenguajes como recursos que enriquecen las configuraciones de prctica matemtica, en sintona con las exigencias de la investigacin y la prctica profesional.



En el mbito internacional, organizaciones como el **NCTM (2000, 2014)** han subrayado que la tecnologa es esencial para la enseanza y el aprendizaje de la matemtica, dado que facilita la experimentacin, la validacin de conjeturas y la comunicacin de ideas. Asimismo, instancias como la **ACM** y el **IEEE** han sealado la importancia de desarrollar habilidades computacionales como requisito fundamental en la formacin universitaria, tanto en el sector pblico como privado.



Por tanto, ensear matemticas integrando Python, LaTeX, Markdown e ingls en ambientes como Jupyter Notebooks contribuye a la formacin de profesionales con **competencias cientficas, comunicativas y tecnolgicas**, capaces de comprender la matemtica de manera interactiva y experimental, y de responder a las demandas del mundo laboral. Esta integracin, lejos de ser inadecuada, constituye un **puente entre la enseanza acadmica y las necesidades de la sociedad**, fortaleciendo la pertinencia de los programas universitarios.

# Estado del arte

El inters por integrar la tecnologa en la enseanza de las matemticas tiene races en las propuestas de **Papert (1980, 1996)**, quien con su enfoque constructionista concibi la computadora como un objeto para pensar, capaz de mediar la construccin activa de significados matemticos mediante la programacin. Esta perspectiva se enlaza con los planteamientos de **Piaget (1970)** y **Vygotsky (1978)**, quienes reconocieron la importancia de la experimentacin y de las herramientas culturales en la formacin del pensamiento. Bajo esta mirada, el uso de entornos digitales no constituye una amenaza para la enseanza tradicional de las matemticas, sino un recurso que ampla las posibilidades de aprendizaje.



A comienzos del siglo XXI, la nocin de **pensamiento computacional** introducida por **Wing (2006)** renov el debate al proponer que esta competencia deba ensearse de manera transversal desde los niveles iniciales hasta la educacin superior, situndola al mismo nivel que la lectura, la escritura y las matemticas. 

Este enfoque se alinea con los planteamientos del **NCTM (2000, 2014)**, que destac la tecnologa como un componente esencial para la experimentacin, la representacin y la comunicacin en matemticas. En paralelo, aportes tericos como la **teora de registros de representacin semitica de Duval (1993, 2006)** y el **Enfoque Ontosemitico de Godino (2007, 2017)** muestran que el aprendizaje matemtico se enriquece al transitar entre mltiples sistemas de representacin, incluyendo los digitales.



En el contexto actual, organizaciones como la **ACM** y el **IEEE** insisten en que las competencias computacionales son indispensables en la formacin profesional, lo que refuerza la pertinencia de integrar lenguajes como **Python, LaTeX, Markdown e ingls** en ambientes como **Jupyter Notebooks** para la enseanza universitaria de las matemticas. As, el estado del arte evidencia un consenso creciente: lejos de ser un obstculo, el pensamiento computacional y el uso de lenguajes digitales constituyen una va para fortalecer la comprensin matemtica, formar profesionales ms competentes y responder a las demandas del mundo laboral y cientfico contemporneo.

# Estado del arte con citas



El uso de la tecnologa en la enseanza de las matemticas encuentra uno de sus pilares en **Papert (1980)**, quien afirmaba que *los nios no aprenden sobre la computadora, aprenden con la computadora* (p. 6). En su visin, la programacin se convierte en un medio para la construccin activa de significados matemticos, idea que se enlaza con la epistemologa gentica de **Piaget (1970)**, quien sostuvo que *el conocimiento se construye a travs de la accin* (p. 15), y con la perspectiva de **Vygotsky (1978)**, al sealar que *toda funcin en el desarrollo cultural del nio aparece dos veces: primero en el plano social y despus en el individual* (p. 57). Estos enfoques coinciden en otorgar a las herramientas culturales un papel esencial en la mediacin del aprendizaje.



Posteriormente, **Wing (2006)** acu el trmino pensamiento computacional, definindolo como *una habilidad fundamental para todos, no solo para los cientficos de la computacin* (p. 33). 

Esta autora subray que deba ensearse de manera transversal, en paralelo con la lectura, la escritura y la matemtica. En la misma lnea, el **NCTM (2000)** destac que *la tecnologa es esencial en la enseanza y el aprendizaje de las matemticas; influye en la matemtica que se ensea y mejora el aprendizaje de los estudiantes* (p. 24).



Desde la didctica de las matemticas, **Duval (2006)** mostr que *la comprensin matemtica requiere la coordinacin de al menos dos registros de representacin diferentes* (p. 107), lo que legitima la incorporacin de registros digitales como LaTeX, Markdown y Python. 

Por su parte, **Godino (2017)** plante que *el significado de un objeto matemtico est constituido por el sistema de prcticas en el que interviene* (p. 7), con lo cual la prctica digital se convierte en parte constitutiva del aprendizaje matemtico.



Finalmente, organismos internacionales como la **ACM y el IEEE (2013)** enfatizan que *las habilidades computacionales son esenciales para el ejercicio profesional en cualquier disciplina* (p. 12). 

De esta manera, se consolida un consenso: la enseanza universitaria de las matemticas debe integrar herramientas contemporneas como **Jupyter Notebooks, Python, LaTeX, Markdown e ingls**, no para competir con la enseanza tradicional, sino para **fortalecer la comprensin matemtica, potenciar la comunicacin cientfica y responder a las demandas del mundo laboral**.
