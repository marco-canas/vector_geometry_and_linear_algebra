# Cómo se suman vectores

Dame, una visualización con matplotlib en una arreglo o matriz de 1x2 para mostrar las dos maneras de sumar geométricamente vectores: disposición de los vectores en forma principio fin y disposición de los dos vectores coincidiendo en el origen y trazando el paralelogramo determinado por ellos. 

Aquí tienes la visualización solicitada: una imagen combinada en formato 1×2 que muestra **las dos maneras geométricas de sumar vectores**:

1. **Suma principio–fin (tip-to-tail)**
2. **Suma por paralelogramo**

La figura ya está renderizada arriba. Si deseas guardarla como archivo PNG, modificar colores, vectores, cuadrícula o textos, dímelo y lo ajusto.


````python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

# Create two separate figures and save them temporarily
# Vector definitions
v1 = np.array([2, 1])
v2 = np.array([1, 2])

# First plot: tip-to-tail addition
fig1 = plt.figure()
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(v1[0], v1[1], v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='black')
sum_vec = v1 + v2
plt.quiver(0, 0, sum_vec[0], sum_vec[1], angles='xy', scale_units='xy', scale=1)
plt.xlim(0, 4)
plt.ylim(0, 4)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Principio–fin (tip-to-tail)")
fig1.savefig('/mnt/data/plot1.png')
plt.close(fig1)

# Second plot: parallelogram method
fig2 = plt.figure()
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(v2[0], v2[1], v1[0], v1[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(v1[0], v1[1], v2[0], v2[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, sum_vec[0], sum_vec[1], angles='xy', scale_units='xy', scale=1)
plt.xlim(0, 4)
plt.ylim(0, 4)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Paralelogramo")
fig2.savefig('/mnt/data/plot2.png')
plt.close(fig2)

# Load and concatenate the images horizontally
img1 = Image.open('/mnt/data/plot1.png')
img2 = Image.open('/mnt/data/plot2.png')
combined_width = img1.width + img2.width
combined_height = max(img1.height, img2.height)
combined = Image.new('RGB', (combined_width, combined_height))
combined.paste(img1, (0, 0))
combined.paste(img2, (img1.width, 0))

# Display combined image
plt.figure(figsize=(10, 5))
plt.imshow(combined)
plt.axis('off')

````

# La suma de vectores es conmutativa  


````python
# %% Representación gráfica de la suma de vectores con el método del paralelogramo
import matplotlib.pyplot as plt
import numpy as np

# Definición de los vectores
v1 = np.array([1, 2])
v2 = np.array([-1, 2])

# Vector suma
v_sum = v1 + v2

# Configurar el plano
plt.figure(figsize=(7,7))
plt.axhline( y = 0, color='gray', linewidth=1) # grafica el eje horizontal 
plt.axvline(x = 0, color='gray', linewidth=1)
plt.grid(True, linestyle='--', alpha=0.5)

# Dibujar vectores desde el origen
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', label=r'$\vec{v_1} = (3, 4)$')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='green', label=r'$\vec{v_2} = (2, -1)$')

# Dibujar los lados del paralelogramo
plt.quiver(v1[0], v1[1], v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='lightgreen', alpha=0.6)
plt.quiver(v2[0], v2[1], v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='lightblue', alpha=0.6)

# Dibujar el vector suma (diagonal del paralelogramo)
plt.quiver(0, 0, v_sum[0], v_sum[1], angles='xy', scale_units='xy', scale=1, color='red', label=r'$\vec{v_T} = \vec{v_1} + \vec{v_2}$')

# Ajustar límites del gráfico
plt.xlim(-4, 6)
plt.ylim(-2, 6)

# Etiquetas y título
plt.title("Suma de vectores con el método del paralelogramo", fontsize=13)
plt.xlabel("Eje X (km)")
plt.ylabel("Eje Y (km)")
plt.legend(loc = 'lower right')
plt.savefig(r"C:\Users\marco\Downloads\suma_vectores_paralelogramo.png", dpi=300)
plt.show()
````
