### **Pregunta 1 (10%) ‚Äî Suma de vectores**
Explique con sus propias palabras y mediante un esquema conceptual (dibujado o descrito) qu√© significa sumar dos vectores en el plano.  
Indique:  
a) Propiedades de la suma de vectores.  
b) Interpretaci√≥n geom√©trica del paralelogramo.

---


````python
# %% Representaci√≥n gr√°fica de la suma de vectores con el m√©todo del paralelogramo
import matplotlib.pyplot as plt
import numpy as np

# Definici√≥n de los vectores
v1 = np.array([1, 2])
v2 = np.array([-1, 2])

# Vector suma
v_sum = v1 + v2

# Configurar el plano
plt.figure(figsize=(7,7))
plt.axhline( y = 0, color='gray', linewidth=1) # grafica el eje horizontal 
plt.axvline(x = 0, color='gray', linewidth=1)
plt.grid(True, linestyle='--', alpha=0.5)

# Dibujar vectores desde el origen
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', label=r'$\vec{v_1} = (3, 4)$')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='green', label=r'$\vec{v_2} = (2, -1)$')

# Dibujar los lados del paralelogramo
plt.quiver(v1[0], v1[1], v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='lightgreen', alpha=0.6)
plt.quiver(v2[0], v2[1], v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='lightblue', alpha=0.6)

# Dibujar el vector suma (diagonal del paralelogramo)
plt.quiver(0, 0, v_sum[0], v_sum[1], angles='xy', scale_units='xy', scale=1, color='red', label=r'$\vec{v_T} = \vec{v_1} + \vec{v_2}$')

# Ajustar l√≠mites del gr√°fico
plt.xlim(-4, 6)
plt.ylim(-2, 6)

# Etiquetas y t√≠tulo
plt.title("Suma de vectores con el m√©todo del paralelogramo", fontsize=13)
plt.xlabel("Eje X (km)")
plt.ylabel("Eje Y (km)")
plt.legend(loc = 'lower right')
plt.savefig(r"C:\Users\marco\Downloads\suma_vectores_paralelogramo.png", dpi=300)
plt.show()
````
