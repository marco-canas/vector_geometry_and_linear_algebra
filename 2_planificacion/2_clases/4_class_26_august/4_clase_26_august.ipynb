##  Paso 5: Transpuesta de una matriz

```python

```

 **Explicacin:** La transpuesta consiste en intercambiar filas por columnas.

---



##  Paso 6: Inversa de una matriz



```python
A_inv = np.linalg.inv(A)
print("Inversa de A:\n", A_inv)

# Verificacin: A * A_inv = Identidad
I = np.dot(A, A_inv)
print("A * A_inv:\n", I)
```



 **Explicacin:**

* Una matriz es invertible si es **cuadrada** y su **determinante no es 0**.
* La inversa es la "matriz que deshace" el efecto de la original, tal que:

  $$
  A \cdot A^{-1} = I
  $$

## 10. Retroalimentacin

* Uso de `ipywidgets` para mostrar transformaciones en tiempo real y discutir errores en clase.


````python
import ipywidgets as widgets
from IPython.display import display

def escalar_rotar(alpha, theta):
    import numpy as np
    from math import cos, sin, radians
    A = np.array([[alpha*cos(radians(theta)), -alpha*sin(radians(theta))],
                  [alpha*sin(radians(theta)),  alpha*cos(radians(theta))]])
    return A

widgets.interact(escalar_rotar, alpha=(0.5,2,0.1), theta=(0,360,15));
````
