# Cómo ver los pixeles que forman una imagen y separar las matrices asociadas a cada color de la imagen. 


<img src = 'https://github.com/marco-canas/vector_geometry_and_linear_algebra/blob/main/images/el_grupo_2025-08-05%20at%203.47.50%20PM.jpeg?raw=true'>


Aquí tienes un programa en Python que carga la imagen y muestra los valores numéricos (matriz de píxeles) usando NumPy:


````python
import numpy as np # Libreria experta en vectores y matrices 
from PIL import Image # Libreria para la manipulación de imágenes
import matplotlib.pyplot as plt # Libreria para la visualización de gráficas
````

````python

# Cargar la imagen
imagen = Image.open("el_grupo_2025-08-05 at 3.47.50 PM.jpeg")
````

````python
imagen 
````

````python

# Convertir a matriz de NumPy
matriz = np.array(imagen)
matriz 
````

````python
type(matriz)
````

````python
matriz.dtype
````

````python
matriz.ndim 
````

````python

# Mostrar la forma de la matriz (alto, ancho, canales)
print("La forma de la matriz, es decir, el número de filas y columnas es :",\
    matriz.shape)
````

````python

# Mostrar algunos valores de píxeles
print("Valores de píxeles (una porción):")
print(matriz[100:110, 100:110, 2])  # Cambia los índices para ver otras partes
````

````python

# Mostrar la imagen
plt.imshow(matriz)
plt.axis('off')
plt.title("Imagen cargada")
plt.show()

````

¿Quieres que convierta la imagen a escala de grises, o que exporte la matriz a un archivo `.npy` o `.csv` para analizarla más a fondo?


````python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Componente Roja
axs[0].imshow(matriz[:, :, 0], cmap='Reds')
axs[0].set_title('Componente Roja (R)')
axs[0].axis('off')

# Componente Verde
axs[1].imshow(matriz[:, :, 1], cmap='Greens')
axs[1].set_title('Componente Verde (G)')
axs[1].axis('off')

# Componente Azul
axs[2].imshow(matriz[:, :, 2], cmap='Blues')
axs[2].set_title('Componente Azul (B)')
axs[2].axis('off')

plt.tight_layout()
plt.show()
````
